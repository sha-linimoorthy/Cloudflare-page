```python
from flask import Flask, jsonify, request
import requests

app = Flask(__name__)

API_TOKEN = "ppHIs3xXY5TwPEa2yHhpyqAvppEKFeAbAx1YC_gA"
ACCOUNT_ID = "283da2a44a0d4d85fb642f67e2c283e7"
API_BASE_URL = f"https://api.cloudflare.com/client/v4/accounts/{ACCOUNT_ID}/ai/run/"
headers = {"Authorization": f"Bearer {API_TOKEN}"}

def run(model, input):
    response = requests.post(f"{API_BASE_URL}{model}", headers=headers, json=input)
    return response.json()

@app.route('/process_stories', methods=['POST'])
def process_stories():
    data = request.json

    if 'stories' not in data:
        return jsonify({'error': 'Invalid request. Missing "stories" field'}), 400

    stories = data['stories']
    model = "@cf/baai/bge-base-en-v1.5"

    output = run(model, {"text": stories})
    return jsonify(output)

if __name__ == '__main__':
    app.run(port=5000)
```
